```python
def Matriz_linha(linha):
   
    while True:
        
        linha = input('Digite a Linha:  ')

        if(linha == 'A' or linha == 'a'):
            linha = 0
            break
        elif(linha == 'B' or linha == 'b'):
            linha = 1
            break
        elif(linha == 'C' or linha == 'c'):
            linha = 2
            break
        elif(linha == 'D' or linha == 'd'):
            linha = 3
            break
        elif(linha == 'E' or linha == 'e'):
            linha = 4
            break
        elif(linha == 'F' or linha == 'f'):
            linha = 5
            break
        elif(linha == 'G' or linha == 'g'):
            linha = 6
            break
        elif(linha == 'H' or linha == 'h'):
            linha = 7
            break
        elif(linha == 'I' or linha == 'i'):
            linha = 8
            break
        elif(linha == 'J' or linha == 'j'):
            linha = 9
            break
        elif(linha == 'K' or linha == 'k'):
            linha = 10
            break
        elif(linha == 'L' or linha == 'l'):
            linha = 11
            break
        elif(linha == 'M' or linha == 'm'):
            linha = 12
            break
        elif(linha == 'N' or linha == 'n'):
            linha = 13
            break
        elif(linha == 'O' or linha == 'o'):
            linha = 14
            break
        elif(linha == 'P' or linha == 'p'):
            linha = 15
            break
        elif(linha == 'Q' or linha == 'q'):
            linha = 16
            break
        elif(linha == 'R' or linha == 'r'):
            linha = 17
            break
        elif(linha == 'S' or linha == 's'):
            linha = 18
            break
        elif(linha == 'T' or linha == 't'):
            linha = 19
            break
        elif(linha == 'U' or linha == 'u'):
            linha = 20
            break
        else:
            print ('Digita uma letra certa')

    return linha
```

```python
def Matriz_coluna(coluna,linha):

    if(linha < 5):
        while True:
            coluna = input('Digite as Colunas: ')
            if(int(coluna) < 8):
                break
    elif(linha < 21):
        while True:
            coluna = input('Digite as Colunas: ')
            if(int(coluna) < 20):
                break


    return coluna
```

```python
def Matriz_inteira(A):
    inteira = 0

    for i in range(21):

        for j in range(20):

            if(i < 5 and j == 8):
                break
            if(A[i][j] == 1):    
                inteira += 1
    return inteira
```

```python
def Matriz_meia(A):
    sobra = 0

    for i in range(21):

        for j in range(20):

            if(i < 5 and j == 8):
                break
            if(A[i][j] == 2):
                sobra += 1
                
    return sobra
```

```python
def Matriz_fundo(A):
    assentos = 0

    for i in range(21):

        for j in range(20):

            if(i < 5 and j == 8):
                break
            if(A[i][j] != 0):
                assentos += 1

    return assentos
```
