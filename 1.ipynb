```python
def Matriz_linha(linha):
    linha = input('Digite a Linha:  ')

    if(linha == 'A' or linha == 'a'):
        linha = 0
    elif(linha == 'B' or linha == 'b'):
        linha = 1
    elif(linha == 'C' or linha == 'c'):
        linha = 2
    elif(linha == 'D' or linha == 'd'):
        linha = 3
    elif(linha == 'E' or linha == 'e'):
        linha = 4
    elif(linha == 'F' or linha == 'f'):
        linha = 5
    elif(linha == 'G' or linha == 'g'):
        linha = 6
    elif(linha == 'H' or linha == 'h'):
        linha = 7
    elif(linha == 'I' or linha == 'i'):
        linha = 8
    elif(linha == 'J' or linha == 'j'):
        linha = 9
    elif(linha == 'K' or linha == 'k'):
        linha = 10
    elif(linha == 'L' or linha == 'l'):
        linha = 11
    elif(linha == 'M' or linha == 'm'):
        linha = 12
    elif(linha == 'N' or linha == 'n'):
        linha = 13
    elif(linha == 'O' or linha == 'o'):
        linha = 14
    elif(linha == 'P' or linha == 'p'):
        linha = 15
    elif(linha == 'Q' or linha == 'q'):
        linha = 16
    elif(linha == 'R' or linha == 'r'):
        linha = 17
    elif(linha == 'S' or linha == 's'):
        linha = 18
    elif(linha == 'T' or linha == 't'):
        linha = 19
    elif(linha == 'U' or linha == 'u'):
        linha = 20
    else:
        print ('Digita uma letra certa')

    return linha
```

```python
def Matriz_coluna(coluna,linha):

    if(linha < 5):
        while True:
            coluna = input('Digite as Colunas: ')
            if(int(coluna) < 8):
                break
    elif(linha < 21):
        while True:
            coluna = input('Digite as Colunas: ')
            if(int(coluna) < 20):
                break


    return coluna
```
